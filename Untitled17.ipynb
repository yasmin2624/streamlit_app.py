```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_url = "http://books.toscrape.com/catalogue/page-{}.html"
```

```python
# لتخزين جميع الكتب المستخرجة
all_books = []

# تكرار لجلب الكتب من كل صفحة (على سبيل المثال، أول 5 صفحات)
for page_num in range(1, 51): 
    url = base_url.format(page_num)
    response = requests.get(url)

    # التأكد من نجاح الاتصال
    if response.status_code == 200:
        print(f"Successfully accessed page {page_num}")
    else:
        print(f"Failed to access page {page_num}")
        continue  # في حالة فشل الوصول للصفحة، يتجاهل الصفحة ويستمر

    # تحويل المحتوى إلى BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # استخراج قائمة الكتب
    books = soup.find_all("article", class_="product_pod")

    # استخراج البيانات لكل كتاب وإضافتها إلى قائمة all_books
    for book in books:
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text
        availability = book.find("p", class_="instock availability").text.strip()
        rating = book.p["class"][1]
        # إضافة الكتاب إلى القائمة
        all_books.append([title, price, availability, rating])
```

```python
# طباعة بيانات الكتب الناتجة
print("Title\t\t\tPrice\t\tAvailability\t\tRating")
print("="*70)

for book in all_books:
    print("\t".join(book))
    
```

```python
# حفظ البيانات في ملف CSV
with open("books.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    
    # كتابة رأس الملف
    writer.writerow(["Title", "Price", "Availability", "Rating"])

    # كتابة البيانات لكل كتاب
    for book in all_books:
        writer.writerow(book)
```

```python
# تحقق إذا كان الملف موجود (تم إنشاؤه بنجاح)
try:
    with open("books.csv", "r"):
        print("Data extraction and CSV file creation completed successfully!")
except FileNotFoundError:
    print("File creation failed.")
```

```python
# قراءة الداتا من ملف الـ CSV
df = pd.read_csv("books.csv")

# التأكد من عدم وجود قيم ناقصة
print("\n Nulls in columns:")
print(df.isnull().sum())
```

```python
# التحقق من وجود تكرارات
duplicates = df.duplicated().sum()

if duplicates > 0:
    print(f"\n{duplicates} duplicate row found, will be deleted.")
    df.drop_duplicates(inplace=True)
else:
    print("\n There are no duplicate rows in the data.")
```

```python
# تنضيف السعر: إزالة £ وتحويله لـ float
df['Price'] = df['Price'].str.replace(r'[^\d.]', '', regex=True).astype(float)
```

```python
# تحويل التقييم من كلمات لأرقام
rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}
df['Rating'] = df['Rating'].map(rating_map)

# تحويل الـ Availability لـ Boolean (True لو In stock)
df['Availability'] = df['Availability'].apply(lambda x: "In stock" in x)
 
```

```python
# عرض أول 5 صفوف بعد التنضيف
print("\n Books after cleaning:")
print(df.head())
```

```python
# حفظ البيانات المنضفة في ملف جديد
df.to_csv("cleaned_books.csv", index=False)
print("\nThe cleaned data is saved as (cleaned_books.csv) file ")
```
